```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mean1 = 55
std_dev1 = 10
num_samples = 800

column1_numbers = np.random.normal(mean1, std_dev1, num_samples)
column1_numbers = np.clip(column1_numbers, 30, 120)
column1_numbers = np.round(column1_numbers).astype(int)

mean2 = 18
std_dev = 3

column2_numbers = np.random.normal(mean2, std_dev, num_samples)
column2_numbers = np.clip(column2_numbers, 12, 26)
column2_numbers = np.round(column2_numbers).astype(int)

column3_numbers = np.random.randint(2, size=num_samples)
column3_numbers[column1_numbers > mean1] = 1

data = {'Miles_Per_Week': column1_numbers,
        'Farthest_run': column2_numbers,
        'Qualified_Agege_Marathon': column3_numbers}

df = pd.DataFrame(data)
```

```python
df.head()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Miles_Per_Week'], df['Farthest_run'], c=df['Qualified_Agege_Marathon'])
plt.xlabel('Miles Per Week')
plt.ylabel('Farthest run')
plt.title('Scatter Plot')
plt.colorbar(label='Qualified Agege Marathon')
plt.show()
```

```python
X = df.iloc[:, 0:2]
```

```python
X
```

```python
y = df.iloc[:, 2]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=54, test_size = 0.25)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model_reg0 = SVC(C=0.1)
```

```python
model_reg0.fit(X_train, y_train)
```

```python
model_reg0.score(X_test, y_test)
```

```python
model_reg0 = SVC(C=1)
model_reg0.fit(X_train, y_train)
model_reg0.score(X_test, y_test)
```

```python
model_reg0 = SVC(C=1000)
model_reg0.fit(X_train, y_train)
model_reg0.score(X_test, y_test)
```
